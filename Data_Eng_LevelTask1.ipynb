#### 1. Difference Between Data Engineering and Data Science

Data Engineer
* Extracts data from various data sources.
* Develops data architecture.
* Builds data pipeline and optimizes data for analyses.
* Makes use of cloud technology.

Data Scientist
* Clean outliers off datasets and analyze them.
* Looks for patterns in datasets and build models usind machine learning techniques.
* Monitor business processes with data.

#### 2. Why I prefer data engineering.

As important as data is to an organisation, the result of the analysis that enables informed-decision making solely depends on how well the data structure and organisation is. I prefer data engineering because I enjoy working with data, optimizing the databases for analysis, and structuring them in their right formats - cleaning, manipulation, tidying. I love the part where I get to work data (small or large) using various medium to improve the system for the analysts and/or scientists. I love being in that part where the real work is done and the impact is felt in the organisation.

#### 5a. Difference between data warehouses and data lakes

Data warehouses - This is mainly used by data analysts, scientists, business analysts, and machine learning engineers.
They mainly store structured data and optimized for analysis. They usually store smaller amount of data.

Data lakes - This is mainly used by data scientists and data engineers. They usually contain unstructured data and there is no need for analysis optimization. They usually store larger amount of data and big data.

#### 5b. Difference between ETL and ELT pipelines.

Where and When the Tranformations happen:
- ETL: transformations happen within the data pipeline.
- ELT: transformations happen at the destination environment.

Flexibilty
- ETL: is rigid. Pipelines are engineered to users specifications.
- ELT: is less rigid. End users build their own transformation.

Time-to-Insight
- ETL: workflows take time to specify and develop.
- ELT: supports self-serve, interactive analytics in real-time.

Schema
- ETL: the conventional approach is schema-on-write. This is consistent, efficient, and has limited versatility.
- ELT: the conventional approach is schema-on-read. This has more versatility and enhanced storage flexibility (more data).
